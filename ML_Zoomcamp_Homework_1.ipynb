```python
# Question 3: What's the average price of BMW cars in the dataset?

df.groupby(by='Make')[['MSRP']].mean().loc['BMW']
```

```python
# Question 4: 
# Select a subset of cars after year 2015. How many of them have missing values for Engine HP?

df[df['Year'] > 2014][['Engine HP']].isna().sum()
```

```python
# Question 5:
# Calculate the average "Engine HP" in the dataset.
# Use the fillna method and to fill the missing values in "Engine HP" 
# with the mean value from the previous step.
# Now, calcualte the average of "Engine HP" again.

mean_hp_before = df['Engine HP'].mean()
mean_hp_after = df['Engine HP'].fillna(mean_hp_before).mean()

round(mean_hp_before, ndigits=13), round(mean_hp_after, ndigits=13), mean_hp_before == mean_hp_after
```

```python
# Select all the "Rolls-Royce" cars from the dataset
# Select only columns "Engine HP", "Engine Cylinders", "highway MPG"
# Now drop all duplicated rows using drop_duplicates method (you should get a dataframe with 7 rows)

df[df['Make'] == 'Rolls-Royce'][["Engine HP", "Engine Cylinders", "highway MPG"]].drop_duplicates()
```

```python
# Get the underlying NumPy array. Let's call it X.
# Compute matrix-matrix multiplication between the transpose of X and X. 
# To get the transpose, use X.T. Let's call the result XTX.
# Invert XTX:

X = df[df['Make'] == 'Rolls-Royce'][["Engine HP", "Engine Cylinders", "highway MPG"]].drop_duplicates().to_numpy()
XTX = np.matmul(X.T, X)
Inv = np.linalg.inv(XTX)
Inv
```

```python
# Question 6:
# What's the sum of all the elements of the result?

np.sum(Inv)
```

```python
# Create an array y with values [1000, 1100, 900, 1200, 1000, 850, 1300].
# Multiply the inverse of XTX with the transpose of X, 
# and then multiply the result by y. Call the result w.

y = [1000, 1100, 900, 1200, 1000, 850, 1300]
w = np.matmul(Inv, X.T).dot(y)
w
```
